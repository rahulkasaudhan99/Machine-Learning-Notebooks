### Stopwords

```python
from nltk.corpus import stopwords
```

```python
sw=set(stopwords.words('english'))
```

```python
print(sw)
```

```python
#function to remove stopwords from a text
def remove_stopwords(text,stopwords):
    useful_words=[w for w in text if w not in stopwords]
    return useful_words
```

```python
text="i am not bothered about her very much"
texts=text.split()    #list of words i.e tokenization
useful_text=remove_stopwords(texts,sw)
```

```python
print(useful_text)
```

```python
'not' in sw
```

Here it also removed 'not' that means it will affect negation in our results. So we have to generate or modify stopwords list according to our need or depending upon use case.

# More Ways to create features
- Unigrams : Every word as a feature
- Bigrams : Two conjuctive word can be treated as a single word
- Tirgrams : 
- n-grams : 
- TF - IDF Normalisation

```python
sent1=["this is good movie"]    #positive review
sent3=["this is good movie but actor was not present"]    #+ve review but based on unigram classified as -ve
sent2=["this is not good movie"] #-ve review
```

- In unigram each unique word is a feature
- But if we use bigram "not good" is feature "a good" is featre
- We can capture Negation "not" word also

```python
cv=CountVectorizer()
```

```python
docs=[sent1[0],sent2[0]]
cv.fit_transform(docs).toarray()
```

here in both the array there is 'not' word in 2nd but not in 1st arr 
- we can make ngram when object created CountVectorizer we can sepicfy our own which is by default (1,1)

```python
#bigram
cv=CountVectorizer(ngram_range=(2,2))    #by default(1,1)
docs=[sent1[0],sent2[0]]
cv.fit_transform(docs).toarray()
```

```python
cv.vocabulary_
```

```python
#trigram
cv1=CountVectorizer(ngram_range=(3,3))    #by default(1,1)
docs=[sent1[0],sent2[0]]
cv1.fit_transform(docs).toarray()
```

```python
cv1.vocabulary_
```

```python
#ngram
cv2=CountVectorizer(ngram_range=(1,3))    #by default(1,1)
docs=[sent1[0],sent2[0]]
cv2.fit_transform(docs).toarray()
```

```python
cv2.vocabulary_
#we can see here vector length also gets increased
```

# Tf-idf Normalisation
- Avoid features that occur very often, bcoz they contain very less information
- Information Decreases as the no of occurences increases across different type of document
- So we define another term - "term-document-frequency"(tdf) which associates a weight with every term
- tf . idf is basically product of two terms  
    1. first term  term freq denoted as - tf (t,d) of term t and doc d
    2. second term is - idf (inverse document frq) denoted as - idf (t,d) =  log(N/(1 + Count(t,D) )) 
    where N=no of docs , (D=all docs) and Count(t,D) count of term t across all docs (i.e how many times t appeared                 across all docs) 
    

-  **for term 'good' the value of tf.idf=1*log(3/3) = nearly 0    here less info (wt will be 0 and 1)** 
-  **for term 'not' the value of tf.idf=1*log(3/1) =  heigher wt    means more info**

```python
sent1="this is good movie"
sent2="this was good movie"
sent3="this is not good movie"
corpus=[sent1,sent2,sent3]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
```

```python
tfidf=TfidfVectorizer()
```

```python
vc=tfidf.fit_transform(corpus).toarray()
print(vc)
```

**here term 'not' has tf.ifd value=0.61 term 'was' has val=0.69 which is higher**
